```python
import numpy as np
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
import matplotlib.pyplot as plt

from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D  
from matplotlib.ticker import LinearLocator
from matplotlib.colors import TwoSlopeNorm
```

```python
sigma=0.0;

x_val=np.pi/2-0.1

z_val=0.5;
k=1;
n=2;
j=1;

c1=0;
c2=1;
c3=0;
c4=1
```

```python
def realpart_w_efunc(x,y,z):
    return np.exp((sigma/2)*x)*(c1*(np.cos(k*x)*np.cos(n*y))-c2*(np.sin(k*x)*np.sin(n*y)))*np.sin(j*np.pi*z_val)

def impart_w_efunc(x,y,z):
    return np.exp((sigma/2)*x)*(c3*(np.cos(k*x)*np.sin(n*y))+c4*(np.sin(k*x)*np.cos(n*y)))*np.sin(j*np.pi*z_val)
    
    
   # np.exp((sigma/2)*x)*np.sin(j*np.pi*z_val)*h(x)*g(y)
    
    #np.exp((sigma/2)*x)*(np.cos(x)*np.cos(y)-np.sin(x)*np.sin(y))*np.sin(np.pi*z_val)
```

```python
X=np.linspace(-np.pi,np.pi,100)
Y=np.linspace(-2*np.pi,2*np.pi,100)
Z=np.linspace(0,5,100)

X, Y = np.meshgrid(X, Y)

real_w_vals=realpart_w_efunc(X,Y,z_val)

im_w_vals=impart_w_efunc(X,Y,z_val)
```

```python
fig, ax = plt.subplots()
ax.tick_params(axis="both", which="major", labelsize=18)  
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$y$",fontsize=18)

plot=ax.contourf(X, Y, real_w_vals, cmap=cm.coolwarm,norm=TwoSlopeNorm(0))#, levels=levels)



#ax.set_ylabel(r"$\alpha_y$",fontsize=20)
#ax.set_xlabel(r"$\beta_y$",fontsize=20)
##ax.contourf(X, Y, Z, cmap=cm.coolwarm)
           
# Add a color bar which maps values to colors.
fig.colorbar(plot)


#ax.set_box_aspect(aspect=(1,1,1))
plt.savefig("efunc_contourplot_sigma0.0.pdf",format='pdf',bbox_inches='tight')
plt.show()
```
