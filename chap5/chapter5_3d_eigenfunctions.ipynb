```python
import pickle
import numpy as np
import matplotlib.colors as mcolors
```

```python
#notebook to find full eigenfunctions

#import streamwise and transverse eigenfunctions

#mulipy them together
```

```python
with open('streamwise_efunc.pck', 'rb') as file_handle:
   streamwise_efunc = pickle.load(file_handle)

with open('transverse_efunc.pck', 'rb') as file_handle:
   transverse_efunc = pickle.load(file_handle)
```

```python
L=100
num_plot_vals=400
plotting_range=np.linspace(-L,L,num_plot_vals)

alpha_x=0.05;
alpha_y=1;

beta_y=-.5;
```

```python
##multiply em together and visualize###

full_efunc=np.empty([num_plot_vals,num_plot_vals])

for i in range(num_plot_vals):
    for j in range(num_plot_vals):
        full_efunc[i][j]=streamwise_efunc[j]*transverse_efunc[i]
```

```python
X,Y=np.meshgrid(plotting_range, plotting_range)
Z=full_efunc
```

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,8))

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
#ax.set_zlim(-1.01, 1.01)
#ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')
# Add a color bar which maps values to colors.
#fig.colorbar(surf, shrink=0.5, aspect=10)
ax.set_ylabel(r"$y$",fontsize=20)
ax.set_xlabel(r"$z$",fontsize=20)
#ax.set_zlabel(r"$p(z,y)$",fontsize=20)

#ax.set_zlim(-.5, .5)
#ax.set_xlim(-50, 50)
#ax.set_ylim(-50, 50)
#ax.set_aspect("equalxy")
#ax.azim = -90 #to see eigenfunction in x direciton

ax.azim = 0 #to see eigenfunction in y direciton

ax.dist = 10
ax.elev = 10

plt.savefig("2devals_view3_nobar.pdf",format='pdf',bbox_inches='tight')

plt.show()
```
